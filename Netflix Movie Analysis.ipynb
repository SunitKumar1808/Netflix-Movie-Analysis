```python
sns.countplot(data=df,x='type')
plt.title('Distribution of Movies vs TV Shows')
plt.show()
```

```python
sns.histplot(df['release_year'],kde=False,bins=30)
plt.title('Content Released by Year')
plt.show()
```

```python
type_counts=df['type'].value_counts()
plt.pie(type_counts,labels=type_counts.index,autopct='%1.1f%%',startangle=90)
plt.title('Movies vs TV Shows Distraction')
plt.show()
```

```python
top_countries=df['country'].value_counts().head(5)
sns.barplot(x=top_countries.values,y=top_countries.index)
plt.title('Top 5 Countries Producing Content')
plt.show()
```

```python
sns.countplot(data=df,y='rating',order=df['rating'].value_counts().index)
plt.title('Distribution of Ratings')
plt.show()
```

```python
sns.histplot(df['release_year'],bins=20,kde=True)
plt.title('Content Released Over the Years')
plt.show()
```
